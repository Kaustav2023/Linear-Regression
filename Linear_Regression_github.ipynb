# Inter Quartile Range (IQR)

```python
df.info()
```

```python
df.columns
```

```python
data=df.loc[:,['symboling', 'wheelbase', 'carlength',
       'carwidth', 'carheight', 'curbweight',
       'enginesize', 'boreratio', 'stroke', 'compressionratio',
       'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']]
```

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
print(IQR)
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
```

```python
df2=data[~(data<LL)|(data>UL).any(axis=1)]
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,5,1)
df2.horsepower.plot(kind='box')

plt.subplot(1,5,2)
df2.peakrpm.plot(kind='box')

plt.subplot(1,5,3)
df2.citympg.plot(kind='box')

plt.subplot(1,5,4)
df2.highwaympg.plot(kind='box')

plt.subplot(1,5,5)
df2.price.plot(kind='box')
plt.show()
```

# Mulitcollinearity
- In statistics, multicollinearity is a situation where the predictors in a regression model are linearly dependent.
- Perfect collinearity refers to a situation where the predictive variables have an exact linear relationship. This phenomenon is called <b>data <b>collision

# Varaince Inflation Factor (VIF)
We are selecting columns that are highly correlated and have numerical values.

```python
df2.info()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables=df_normal[df_normal.columns]
```

```python
variables
```

```python
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['columns']=variables.columns
```
